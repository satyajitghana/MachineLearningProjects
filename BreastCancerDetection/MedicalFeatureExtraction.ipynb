# Medical Feature Extraction

```python
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
import numpy as np
```

```python
img = mpimg.imread('IMG-0001-00001.jpg')
plt.figure(figsize=(15, 15))
plt.imshow(img)
plt.show()
```

```python
lum_img = img[:,:,0]
plt.figure(figsize=(15, 15))
plt.imshow(lum_img)
plt.show()
```

```python
plt.figure(figsize=(15, 15))
plt.imshow(lum_img, cmap='hot')
plt.colorbar()
plt.show()
```

```python
plt.figure(figsize=(15, 15))
plt.imshow(lum_img, clim=(0.0, 0.7))
plt.show()
```

```python
plt.figure(figsize=(15, 15))
plt.imshow(lum_img, interpolation='bicubic')
plt.show()
```
