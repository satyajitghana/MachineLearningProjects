# Fashion Items Classification

```python
import keras
import matplotlib.pyplot as plt
import numpy as np
from keras.datasets import fashion_mnist
```

# Load the Data

```python
(x_train, y_train), (x_test, y_test) = fashion_mnist.load_data()
(x_train.shape, y_train.shape), (x_test.shape, y_test.shape)
```

## Let's look at the test and the sample data 

Label	Description
0. T-shirt/top
1. Trouser
2. Pullover
3. Dress
4. Coat
5. Sandal
6. Shirt
7. Sneaker
8. Bag
9. Ankle boot

```python
labels = ['T-Shirt/Top', 'Trouser', 'Pullover', 'Dress', 'Coat', 'Sandal', 'Shirt', 'Sneakers', 'Bag', 'Ankle Boot']
```

```python
n = 25 # first 20 images
plt.figure(figsize=(15, 15))
for i in range(n):
    plt.subplot(5, 5, i+1)
    plt.xticks([])
    plt.yticks([])
    plt.xlabel(labels[y_test[i]])
    plt.imshow(x_test[i])
```

# Normalize the inputs

```python
x_train = x_train / 255.0
x_test = x_test / 255.0
```

# Build the Model

```python
model = keras.models.Sequential()
model.add(keras.layers.Flatten())
model.add(keras.layers.Dense(128, activation='relu'))
model.add(keras.layers.Dense(10, activation='softmax'))
model.compile(optimizer='adam', 
              loss='sparse_categorical_crossentropy',
              metrics=['accuracy'])
model.fit(x_train, y_train, epochs=5, verbose=1)
```

# Summary of the Model

```python
model.summary()
```

# Save and Test the Model

```python
model.save('mnist_fashion.model')
```

```python
old_model = keras.models.load_model('mnist_fashion.model')
test_loss, test_acc = old_model.evaluate(x_test, y_test)
```

```python
print('Loss\t\t: {0}\nAccuracy\t: {1}'.format(test_loss, test_acc))
```

# Predictions

```python
n = 25 # take first 25 images for testing
predicted = old_model.predict(x_test[:n])
predicted.shape
```

```python
plt.figure(figsize=(15, 15))
for i in range(n):
    plt.subplot(5, 5, i+1)
    plt.imshow(x_test[i])
    found = np.argmax(predicted[i])
    plt.xlabel('{0} with Accuracy : {1:.4f}'.format(labels[found], predicted[i][found]))
    plt.ylabel('Actual : {0}'.format(labels[y_test[i]]))
    plt.xticks([])
    plt.yticks([])
```
