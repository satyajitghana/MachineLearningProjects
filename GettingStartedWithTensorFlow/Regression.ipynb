```python
import tensorflow as tf
import numpy as np
import matplotlib.pyplot as plt
```

```python
number_of_points = 500
x_point = []
y_point = []

a = 0.22
b = 0.78

for i in range(number_of_points):
    x = np.random.normal(0.0, 0.5)
    y = a * x + b + np.random.normal(0.0, 0.1)
    x_point.append([x])
    y_point.append([y])
```

```python
plt.figure(figsize=(10, 10))
fig = plt.plot(x_point, y_point, 'o', label='InputData', linewidth=2.0)
# plt.setp(fig, 'color', 'r', 'linewidth', 10.0)
plt.legend()
plt.grid()
plt.title('Data for Linear Regression')
plt.show()
```

# Gradient Descent

```python
a = tf.Variable(tf.random_uniform([1], -1.0, 1.0))
b = tf.Variable(tf.zeros([1]))
y = a * x_point + b

cost_function = tf.reduce_mean(tf.square(y - y_point))
optimizer = tf.train.GradientDescentOptimizer(learning_rate=0.5)
train = optimizer.minimize(cost_function)
```

```python
model = tf.global_variables_initializer()
```

```python
with tf.Session() as sess:
    sess.run(model)
    for step in range(20):
        sess.run(train)
        if (step % 5 == 0):
            plt.plot(x_point, y_point, 'o', label = 'step = {}'.format(step))
            plt.plot(x_point, sess.run(a) * x_point + sess.run(b))
            print('a = ',sess.run(a), '\nb = ', sess.run(b))
            plt.legend()
            plt.show()
```
