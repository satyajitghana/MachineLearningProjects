```python
import matplotlib.pyplot as plt
import pandas as pd
from ggplot import *
```

```python
ggplot(mtcars, aes('mpg', 'qsec')) + geom_point(colour='steelblue') + scale_x_continuous(breaks=[10,20,30],  \
                     labels=["horrible", "ok", "awesome"])
```

```python
p = ggplot(diamonds, aes(x='price'))
p + geom_density() + \
    facet_grid("cut", "clarity")
```

```python
p = ggplot(diamonds, aes(x='carat', y='price'))
p + geom_point(alpha=0.25) + \
    facet_grid("cut", "clarity")
```

```python
gapminder_url='https://bit.ly/2cLzoxH'
gapminder = pd.read_csv(gapminder_url)
gapminder.head()
```

```python
p = ggplot(gapminder) + \
    geom_point(aes(x='gdpPercap', y='lifeExp',color='continent')) + \
    ggtitle('plotnine example: scatter plot')
```

```python
p = ggplot(gapminder, aes(x='gdpPercap', y='lifeExp')) + geom_point(colour='red')
```
