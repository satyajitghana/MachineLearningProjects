# Introduction to Neural Networks

Use the signomid function to map numbers from 0 to 1
$$S(x)=\frac{1}{1+e^{-x}}$$

````python
import numpy as np
# sigmoid function
def sigmoid(x):
    return 1/(1 + np.exp(-x))
# sigmoid prime or derivative of it
def sigmoid_p(x):
    return sigmoid(x) * (1-sigmoid(x))
````

````python
import matplotlib.pyplot as plt
x = np.linspace(-15,15,1000)
plt.plot(x, sigmoid(x), 'c', lw=2)
plt.title('The Sigmoid Function')
plt.xlabel('X')
plt.ylabel('Y')
plt.grid()
plt.show()
````

$$NN(m1,m2)=S(w1\cdot m1+w2\cdot m2+b)$$
Where $m1, m2$ are the inputs, $w1, w2$ are the weights and $b$ is the biasing factor, everything is squished down using the sigmoid function to give an output from 0 to 1.
To minimize the error, Method of Least Squares can be used, and then any form of regression to fit the curve, this will provide the correct amount of weights for the best curve fit

#### Training with a very minimal sample data

````python
data = [[3,   1.5,  1],
        [2,   1,    0],
        [4,   1.5,  1],
        [3,   1,    0],
        [3.5, 0.5,  1],
        [2,   0.5,  0],
        [5.5, 1,    1],
        [1,   1,    0]]
print(data)
````

```
  network
    o - flower type
   / \  w1, w2
  o   o  length, width```

````python
w1 = np.random.randn()
w2 = np.random.randn()
b  = np.random.randn()
````

````python
X = np.linspace(-10, 10, 100)
plt.plot(X, sigmoid(X), 'r')
plt.plot(X, sigmoid_p(X), 'c')
plt.show()
````

````python
# scatter data
plt.grid()
for i in range(len(data)):
    point = data[i]
    color = "r"
    if point[2] == 0:
        color = "c"
    plt.scatter(point[0], point[1], c=color)
plt.show()
````

````python
# training loop

learning_rate = 0.2
costs = []

for i in range(50000):
    ri = np.random.randint(len(data))
    point = data[ri]
    
    z = point[0] * w1 + point[1] * w2 + b
    pred = sigmoid(z)
    target = point[2]
    cost = np.square(pred - target)
    
    costs.append(cost)
    
    dcost_pred = 2 * (pred - target)
    dpred_dz = sigmoid_p(z)
    
    dz_dw1 = point[0]
    dz_dw2 = point[1]
    dz_db  = 1
    
    # partial derivatives
    
    dcost_dz  = dcost_pred * dpred_dz
    
    dcost_dw1 = dcost_dz * dz_dw1
    dcost_dw2 = dcost_dz * dz_dw2
    dcost_db  = dcost_dz * dz_db
    
    w1 = w1 - learning_rate * dcost_dw1
    w2 = w2 - learning_rate * dcost_dw2
    b  = b  - learning_rate * dcost_db
plt.plot(costs)
plt.show()
````

````python
# predicting the model

for i in range(len(data)):
    point = data[i]
    print(point)
    z = point[0] * w1 + point[1] * w2 + b
    pred = sigmoid(z)
    print(pred)
````

````python
def which_flower(length, width):
    z = length * w1 + width * w2 + b
    pred = sigmoid(z)
    print(pred)
    if pred < 0.5:
        print("blue")
    else:
        print("red")
````

````python
which_flower(5.5 ,1)
````

````python
# check out the networks predictions in the x,y plane
for x in np.linspace(0, 6, 20):
    for y in np.linspace(0, 3, 20):
        pred = sigmoid(w1 * x + w2 * y + b)
        c = 'c'
        if pred > .5:
            c = 'r'
        plt.scatter([x],[y],c=c, alpha=.2)
        
# plot points over network predictions
# you should see a split, with half the predictions blue
# and the other half red.. nicely predicting each data point!

# scatter data
plt.grid()
for i in range(len(data)):
    point = data[i]
    color = "r"
    if point[2] == 0:
        color = "c"
    plt.scatter(point[0], point[1], c=color)
plt.show()
````

````python
data = np.array(data)
x = data[:,[0,1]]
y = data[:,2]
# xsquare = np.multiply(x, x)
# n = np.ones((1, len(data)))
# S = np.array([x, xsquare, n])
# print(S)
# A = np.dot(S, np.transpose(S))
# print(A)
# B = np.dot(S, np.transpose(y))
# #S
# C = np.polyfit(x, y, 1)
# print(C[0])
# prediction = C[0]
# def prediction(x, y):
#     return C[0]*x+C[1]
from sklearn import linear_model
# regression = linear_model.LinearRegression()
regression = linear_model.Ridge (alpha = .5)
regression.fit(x, y)
for i in range(len(data)):
    print(data[i, :])
    print(regression.predict([[data[i, 0], data[i, 1]]]))
    if regression.predict([[data[i, 0], data[i, 1]]]) <= 0.5 :
        print("Red")
    else :
        print("Blue")
        
# check out the networks predictions in the x,y plane
for x in np.linspace(0, 6, 20):
    for y in np.linspace(0, 3, 20):
        pred = regression.predict([[x, y]])
        c = 'c'
        if pred > .5:
            c = 'r'
        plt.scatter([x],[y],c=c, alpha=.2)
        
# plot points over network predictions
# you should see a split, with half the predictions blue
# and the other half red.. nicely predicting each data point!

# scatter data
plt.grid()
for i in range(len(data)):
    point = data[i]
    color = "r"
    if point[2] == 0:
        color = "c"
    plt.scatter(point[0], point[1], c=color)
plt.show()
````
